```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
df = pd.read_excel('data/cxb-bangladesh-barriers-data_040419.xlsx')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

### Washington Group Questions
1. Do you have difficulty seeing even if wearing glasses?
2. Do you have difficulty hearing, even if using a hearing aid?
3. Do you have difficulty walking or climbing steps?
4. Do you have difficulty remembering or concentrating?
5. Do you have difficulty washing all over or dressing?
6. Using your usual language, do you have difficulty communicating?

### Sectors
- Health
- Shelter
- Wash
- Food Distribution
- Protection
- Education
- Access to Livelihood

```python
df.gender.value_counts().plot(kind='bar');
```

```python
df.age.value_counts().plot(kind='bar');
```

```python
mask = (df.cohabitating_with_spouse=='yes')
df.loc[mask, 'content_rating'] = 'NC-17'
df.groupby('married').head
```

```python
sns.pairplot(df, x_vars=['WGQ1sight', 'WGQ2hear', 'WGQ3walk', 'WGQ4remember', 'WGQ5selfcare', 'WGQ6speech'], y_vars=['gender', 'age']);

```

```python
df.describe()
```

```python
graph = df[['WGQ1sight', 'WGQ2hear', 'WGQ3walk', 'WGQ4remember', 'WGQ5selfcare', 'WGQ6speech']].mean().sort_values().plot(kind='bar');
plt.title('Washington Group Question Averages');
graph.text(x=-0.5, y=-1.5, s = 'Source: HDX',fontsize = 10, color = '#000000');
```

```python
sns.catplot(x='gender', y='WGQ3walk', kind='swarm', data=df);
```

```python
sns.catplot(x='age', y='WGQ3walk', kind='swarm', data=df, order=['06-11 y', '12-18 y', '19-35 y', '36-50 y', '50+ y']);
```
