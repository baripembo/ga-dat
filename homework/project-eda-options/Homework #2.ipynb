### Washington Group Questions
HI utilizied the Washington Group Questions (short set), which are a validated and endorsed set of questions developed to identify people who have difficulties in basic, universal activities and are thus at risk of restricted participation in an unaccommodating environment.

Q1 - Do you have difficulty seeing even if wearing glasses?  
Q2 - Do you have difficulty hearing, even if using a hearing aid?  
Q3 - Do you have difficulty walking or climbing steps?  
Q4 - Do you have difficulty remembering or concentrating?  
Q5 - Do you have difficulty washing all over or dressing?  
Q6 - Using your usual language, do you have difficulty communicating?  

#### Answers on a scale 1-4
1 - No, no difficulty  
2 - Yes, some difficulty  
3 - Yes, a lot of difficulty  
4 - Cannot do it at all  

```python
sns.pairplot(df, x_vars=['WGQ1sight', 'WGQ2hear', 'WGQ3walk', 'WGQ4remember', 'WGQ5selfcare', 'WGQ6speech'], y_vars=['gender', 'age']);
```

```python
graph = df[['WGQ1sight', 'WGQ2hear', 'WGQ3walk', 'WGQ4remember', 'WGQ5selfcare', 'WGQ6speech']].mean().sort_values().plot(kind='bar');
plt.title('Washington Group Question Averages');
graph.text(x=-0.5, y=-1.5, s = 'Source: HDX',fontsize = 10, color = '#000000');
```
